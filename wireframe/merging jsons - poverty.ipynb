```python
import csv 

#add h05 data
f = open('data/h05_poverty.csv', 'r')
reader = csv.reader(f)
h05 = {}
for i, row in enumerate(reader):
    if i>0:
        h05[row[0]] = row[4]
 
f.close()
h05
```

```python
#add h10 data
g = open('data/h10_poverty.csv', 'r')
reader = csv.reader(g)
h10 = {}
for i, row in enumerate(reader):
    if i>0:
        h10[row[0]] = row[4]

g.close()
h10
```

```python
#add r05 data
h = open('data/r05_poverty.csv', 'r')
reader = csv.reader(h)
r05 = {}
for i, row in enumerate(reader):
    if i>0:
        r05[row[0]] = row[4]

h.close()
r05
```

```python
#add r10 data
j = open('data/r10_poverty.csv', 'r')
reader = csv.reader(j)
r10 = {}
for i, row in enumerate(reader):
    if i>0:
        r10[row[0]] = row[4]

j.close()
r10
```

```python
import json

with open('pumas_pa05.json') as json_data:
    d = json.load(json_data)
#     print(d['objects']['pumas_pa_only05']['geometries'])
#     geo_puma = {}
    
    for i, props in enumerate(d['objects']['pumas_pa_only05']['geometries']): 
        curr_id = props['properties']['id']
        props['properties']['h05_pov'] = h05[curr_id]
        props['properties']['h10_pov'] = h10[curr_id]
        props['properties']['r05_pov'] = r05[curr_id]
        props['properties']['r10_pov'] = r10[curr_id]
        d['objects']['pumas_pa_only05']['geometries'][2][i] = props['properties']
#         print(props['properties'])
    
#     print(d['objects']['pumas_pa_only05'])

# output json file 
with open('pa05_pov.json', 'w') as outfile:
    json.dump(d, outfile)
```
