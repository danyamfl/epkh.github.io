```python
import csv 
 
f = open('data/h05_burden.csv', 'r')
reader = csv.reader(f)
all_pumas = {}
for i, row in enumerate(reader):
    if i>0:
        all_pumas[row[0]] = row[6]
 
f.close()
all_pumas
```
